# **Visualization**

```python
rating_counts = data_after_drop['rating'].value_counts()

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=rating_counts.index, y=rating_counts.values, palette="magma")
plt.title("Distribution of Ratings", fontsize=16)
plt.xlabel("Rating", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

### **The rating distribution indicates the target audience Netflix focuses on. A high count of PG-13 or TV-MA content suggests a preference for teens and adult viewers. Strategic content acquisition can aim to diversify or strengthen specific audience segments.**

```python
movies_df = data_after_drop[data_after_drop['type'] == 'Movie']
movies_df['duration'] = pd.to_numeric(movies_df['duration'].str.replace(' min', ''), errors='coerce')

# Plot distribution
plt.figure(figsize=(10, 6))
sns.distplot(movies_df['duration'], bins=30, kde=True, color="purple")
plt.title("Duration Distribution of Movies", fontsize=16)
plt.xlabel("Duration (minutes)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

### **Most movies fall within a specific duration range (e.g., 90-120 minutes). This suggests user preferences for medium-length movies, which Netflix could prioritize in its library or original productions.**

```python
tv_shows_df = data_after_drop[data_after_drop['type'] == 'TV Show']
tv_shows_df['seasons'] = pd.to_numeric(tv_shows_df['duration'].str.replace(' Season', '').str.replace('s', ''), errors='coerce')

# Plot
plt.figure(figsize=(10, 6))
sns.histplot(tv_shows_df['seasons'], bins=10, kde=False, color="orange")
plt.title("Distribution of TV Show Seasons", fontsize=16)
plt.xlabel("Number of Seasons", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

### **If the majority have 1-2 seasons, it highlights a tendency for short-format series. This aligns with user demand for binge-worthy, concise series and can guide future series development strategies.**

## **Top 10 Genres in Movies and TV Show**

```python
# Extract and count genres (assuming 'listed_in' column contains genres)
from collections import Counter

genres = data_after_drop['listed_in'].dropna().str.split(', ').sum()
genre_counts = Counter(genres).most_common(10)

# Create DataFrame for visualization
genre_df = pd.DataFrame(genre_counts, columns=['Genre', 'Count'])

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Count', y='Genre', data=genre_df, palette="coolwarm")
plt.title("Top 10 Genres in Movies and TV Shows", fontsize=16)
plt.xlabel("Count", fontsize=12)
plt.ylabel("Genre", fontsize=12)
plt.show()
```

## **Top 10 country wise popular actors casted in different Movies/Shows on Netflix**

```python
def country_trace(country, flag = "movie"):
    data_after_drop["from_us"] = data_after_drop['country'].fillna("").apply(lambda x : 1 if country.lower() in x.lower() else 0)
    small = data_after_drop[data_after_drop["from_us"] == 1]
    if flag == "movie":
        small = small[small["duration"] != ""]
    else:
        small = small[small["season_count"] != ""]
    cast = ", ".join(small['cast'].fillna("")).split(", ")
    tags = Counter(cast).most_common(25)
    tags = [_ for _ in tags if "" != _[0]]

    labels, values = [_[0]+"  " for _ in tags], [_[1] for _ in tags]
    trace = go.Bar(y=labels[::-1], x=values[::-1], orientation="h", name="", marker=dict(color="#a678de"))
    return trace

from plotly.subplots import make_subplots
traces = []
titles = ["United States", "","India","", "United Kingdom", "Canada","", "Spain","", "Japan"]
for title in titles:
    if title != "":
        traces.append(country_trace(title))

fig = make_subplots(rows=2, cols=5, subplot_titles=titles)
fig.add_trace(traces[0], 1,1)
fig.add_trace(traces[1], 1,3)
fig.add_trace(traces[2], 1,5)
fig.add_trace(traces[3], 2,1)
fig.add_trace(traces[4], 2,3)
fig.add_trace(traces[5], 2,5)

fig.update_layout(height=1200, showlegend=False)
fig.show()
```

## **Heatmap of content addition**

```python
# Convert 'date_added' to datetime
data_after_drop['date_added'] = pd.to_datetime(data_after_drop['date_added'], errors='coerce')

# Extract month and year
data_after_drop['month'] = data_after_drop['date_added'].dt.month
data_after_drop['year'] = data_after_drop['date_added'].dt.year

# Count content by year and month
monthly_counts = data_after_drop.groupby(['year', 'month']).size().unstack(fill_value=0)

# Heatmap of content additions
plt.figure(figsize=(12, 8))
sns.heatmap(monthly_counts, cmap="YlGnBu", linewidths=0.5, annot = True)
plt.title("Content Additions by Month and Year", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Year", fontsize=12)
plt.show()
```
